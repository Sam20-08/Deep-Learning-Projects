# Resizing image using pillow

```python
from PIL import Image
```

```python
img=Image.open('/content/dog.jpeg')
img_resized=img.resize((200,200))
```

```python
img_resized.save('dog_resized_img.jpg')
```

```python
plt.imshow(img_resized)
plt.show()
```

# converting rgb to grey scale using opencv

```python
img=cv2.imread('/content/dog.jpeg')
img_grey=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
img.shape
```

```python
img_grey.shape
```

```python
type(img)
```

```python
from google.colab.patches import cv2_imshow
```

```python
# displaying the image

cv2_imshow(img_grey)
```

```python
plt.imshow(img_grey,cmap='gray')
plt.show()
```

```python
#saving the img using cv2

cv2.imwrite('dog_greyscaled_img.jpg',img_grey)
```
